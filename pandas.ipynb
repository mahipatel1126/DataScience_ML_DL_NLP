Pandas-DataFrame And Series

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It
provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object,
while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure
with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
# Series
## A pandas series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
## create  a series from dictionary
data = {'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## create a Dataframe from a dictionary of list

data = {
    'Name': ['Krish','John','Jack'],
    'Age': [25,30,45],
    'City': ['Bnaglore','New york', 'Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a dataframe from a list of dictionaries
data = [
    {'Name': 'Krish', 'Age':32,'City': 'Banglore'},
    {'Name': 'John', 'Age':34,'City': 'New york'},
    {'Name': 'Bappy', 'Age':22,'City': 'Florida'},
    {'Name': 'Jack', 'Age':32,'City': 'America'}
]
df = pd.DataFrame(data)
print(df)
print(type(df))
```
